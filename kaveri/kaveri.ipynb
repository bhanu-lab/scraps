```python
import requests
import json
import time
```

```python
authorization = "LOGIN TO https://kaveri.karnataka.gov.in AND GET THE HEADERS FROM THE NETWORK TAB"
cookie = "LOGIN TO https://kaveri.karnataka.gov.in AND GET THE HEADERS FROM THE NETWORK TAB" 
```

```python
def get_talukas(district_code):
    headers = {
        "_append": "3ccbc2d56099f3f6a380b8d4a8763c5d",
        "accept": "application/json",
        "accept-language": "en-IN,en-GB;q=0.9,en-US;q=0.8,en;q=0.7",
        "authorization": authorization,
        "content-type": "application/json",
        "cookie": cookie,
        "origin": "https://kaveri.karnataka.gov.in",
        "referer": "https://kaveri.karnataka.gov.in/",
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36"
    }
    
    data = {"districtCode": str(district_code)}
    
    try:
        response = requests.post(
            "https://kaveri.karnataka.gov.in/api/GetTalukaAsync",
            headers=headers,
            json=data
        )
        response.raise_for_status()
        return response.json()
    except Exception as e:
        print(f"Error fetching talukas for district {district_code}: {str(e)}")
        return None

district_talukas = {}

for district_code, district_name in districts.items():
    print(f"Fetching talukas for {district_name}...")
    result = get_talukas(district_code)
    
    if result:
        district_talukas[district_name] = result
    
    time.sleep(1)

with open('district_talukas.json', 'w', encoding='utf-8') as f:
    json.dump(district_talukas, f, ensure_ascii=False, indent=2)

print("Done! Results saved to district_talukas.json")

print("\nSample of results:")
for district_name, talukas in list(district_talukas.items())[:3]:
    print(f"\n{district_name}:")
    print(talukas)
```

```python
def get_villages(hobli_code):
    headers = {
        "_append": "3ccbc2d56099f3f6a380b8d4a8763c5d",
        "accept": "application/json",
        "accept-language": "en-IN,en-GB;q=0.9,en-US;q=0.8,en;q=0.7",
        "authorization": authorization,
        "content-type": "application/json",
        "cookie": cookie,
        "origin": "https://kaveri.karnataka.gov.in",
        "referer": "https://kaveri.karnataka.gov.in/",
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36"
    }
    
    data = {"hobliCode": str(hobli_code)}
    
    try:
        response = requests.post(
            "https://kaveri.karnataka.gov.in/api/GetVillageAsync",
            headers=headers,
            json=data
        )

        if response.status_code == 429:
            # print(response.headers)
            # print(response.content)
            # pritn(response.json)
            sleep_time = 5
            print(f'sleeping for {sleep_time} seconds')
            time.sleep(sleep_time)

            response = requests.post(
                "https://kaveri.karnataka.gov.in/api/GetVillageAsync",
                headers=headers,
                json=data
            )

        response.raise_for_status()
        return response.json()
    except Exception as e:
        print(f"Error fetching villages for hobli {hobli_code}: {str(e)}")
        return None


i = 0
for hobli_code, hobli_name in hoblis.items():
    if hobli_name in hobli_villages:
        # print(f"skipping villages for hobli {hobli_code}, {hobli_name}, i = {i}")
        i += 1
        continue
    print(f"getting villages for hobli {hobli_code}, {hobli_name}, i = {i}")
    result = get_villages(hobli_code)
    print(result)
    if result:
        hobli_villages[hobli_name] = result
    i += 1
    # time.sleep(1)

with open('hobli_villages.json', 'w', encoding='utf-8') as f:
    json.dump(hobli_villages, f, ensure_ascii=False, indent=2)
```

```python
# Load all the JSON files
with open('district_talukas.json', 'r', encoding='utf-8') as f:
    district_talukas = json.load(f)

with open('taluk_hoblis.json', 'r', encoding='utf-8') as f:
    taluk_hoblis = json.load(f)

with open('hobli_villages.json', 'r', encoding='utf-8') as f:
    hobli_villages = json.load(f)

# Create the reverse mapping
village_mapping = []

seen = set()

# Iterate through the hierarchy to build the mapping
for district_name, talukas in district_talukas.items():
    for taluka in talukas:
        taluk_name = taluka['talukNamee']
        
        # Get hoblis for this taluka
        hoblis = taluk_hoblis.get(taluk_name, [])
        for hobli in hoblis:
            hobli_name = hobli['hoblinamee']
            
            # Get villages for this hobli
            villages = hobli_villages.get(hobli_name, [])
            for village in villages:
                village_name = village['villagenamee']

                if (district_name, taluk_name, hobli_name, village_name) in seen:
                    continue

                seen.add((district_name, taluk_name, hobli_name, village_name))
                # Create the mapping
                village_mapping.append({
                    'village': village_name,
                    'hoblinamee': hobli_name,
                    'talukNamee': taluk_name,
                    'districtNamee': district_name
                })

print(f"Total unique villages: {len(village_mapping)}")
with open('remap.json', 'w', encoding='utf-8') as f:
    json.dump(village_mapping, f, indent=2, ensure_ascii=False)
```
